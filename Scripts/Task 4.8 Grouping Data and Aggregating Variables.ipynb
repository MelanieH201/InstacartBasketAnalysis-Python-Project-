```python
# Importing libraries
import pandas as pd
import numpy as np
import os
```

```python
# Import Data
path = r'/Users/melanie/Desktop/7:2022 Instacart Basket Analysis'
```

```python
ords_prods_merge = pd.read_pickle(os.path.join(path, 'Data', 'Prepared Data', 'orders_products_renamed2.pkl'))
```

```python
df = ords_prods_merge
```

```python
df.shape
```

```python
df.head(10)
```

```python
# Grouping data and aggregating df dataframe
df.groupby('department_id').agg({'order_number':['mean']})
```

The subset mean ranges from 14 to 24, and the entire dataframe mean ranges from 14 to 22. 

```python
# Aggregating data with transform() to create a loyalty flag
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(100)
```

```python
# Deriving with loc() to create a loyalty flag
pd.options.display.max_rows = None
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Finding if price differs between loyalty categories
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

The average price does vary between the loyalty catagories.

```python
# Create a spending flag
ords_prods_merge.loc[ords_prods_merge['prices'] < 10 ,'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[(ords_prods_merge['prices'] <= 10) & (ords_prods_merge['prices'] >= 10), 'spending_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['prices'] >= 10 ,'spending_flag'] = 'High spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Create an order frequency flag
ords_prods_merge['median_order_duration'] = ords_prods_merge.groupby(['user_id'])['days_since_previous_order'].transform(np.median)
```

```python
ords_prods_merge[['user_id', 'days_since_previous_order', 'median_order_duration']].head()
```

```python
ords_prods_merge.loc[ords_prods_merge['median_order_duration'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_order_duration'] <= 20) & (ords_prods_merge['median_order_duration'] > 10), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_order_duration'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge[['user_id', 'median_order_duration', 'frequency_flag']].head(50)
```

```python
ords_prods_merge.loc[ords_prods_merge['median_order_duration'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_order_duration'] > 10) & (ords_prods_merge['median_order_duration'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_order_duration'] < 10, 'frequency_flag'] = 'Frequent customer'
```
